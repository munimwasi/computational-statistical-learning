```python
import statsmodels.api as sm

# Define a function to calculate confidence intervals
def calculate_confidence_intervals(x, y):
    x_with_const = sm.add_constant(x)

    model = sm.OLS(y, x_with_const).fit()

    confidence_intervals = model.conf_int(alpha=0.05)  # 95% confidence intervals

    return confidence_intervals

conf_intervals_original = calculate_confidence_intervals(x, y)

eps_noisier = np.random.normal(0, 0.5, 100)  # Increase variance for more noise
y_noisier = -1 + 0.5 * x + eps_noisier
conf_intervals_noisier = calculate_confidence_intervals(x, y_noisier)

eps_less_noisy = np.random.normal(0, 0.1, 100)  # Decrease variance for less noise
y_less_noisy = -1 + 0.5 * x + eps_less_noisy
conf_intervals_less_noisy = calculate_confidence_intervals(x, y_less_noisy)

print("Confidence Intervals (Original Data):")
print(conf_intervals_original)

print("\nConfidence Intervals (Noisier Data):")
print(conf_intervals_noisier)

print("\nConfidence Intervals (Less Noisy Data):")
print(conf_intervals_less_noisy)
```

```python
############################ PROBLEM 14 #####################

rng = np.random.default_rng(10)
x1 = rng.uniform(0, 1, size=100)
x2 = 0.5 * x1 + rng.normal(size=100) / 10
y = 2 + 2 * x1 + 0.3 * x2 + rng.normal(size=100)
```

```python
import matplotlib.pyplot as plt

correlation = np.corrcoef(x1, x2)[0, 1]
print("Correlation between x1 and x2:", correlation)

plt.scatter(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Scatterplot of x1 vs. x2')
plt.show()
```

```python
import statsmodels.api as sm

X = sm.add_constant(np.column_stack((x1, x2)))
model = sm.OLS(y, X).fit()
coefficients = model.params
print("βˆ0:", coefficients[0])
print("βˆ1:", coefficients[1])
print("βˆ2:", coefficients[2])

# Testing null hypotheses
p_value_beta1 = model.pvalues[1]
p_value_beta2 = model.pvalues[2]
print("p-value for H0: β1 = 0:", p_value_beta1)
print("p-value for H0: β2 = 0:", p_value_beta2)
```

```python
X1_only = sm.add_constant(x1)
model_x1_only = sm.OLS(y, X1_only).fit()
p_value_x1 = model_x1_only.pvalues[1]
print("βˆ0 (x1 only):", model_x1_only.params[0])
print("βˆ1 (x1 only):", model_x1_only.params[1])
print("p-value for H0: β1 (x1 only) = 0:", p_value_x1)
```

```python
X2_only = sm.add_constant(x2)
model_x2_only = sm.OLS(y, X2_only).fit()
p_value_x2 = model_x2_only.pvalues[1]
print("βˆ0 (x2 only):", model_x2_only.params[0])
print("βˆ1 (x2 only):", model_x2_only.params[1])
print("p-value for H0: β1 (x2 only) = 0:", p_value_x2)
```
